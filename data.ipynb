```python
import pandas as pd
import numpy as np
import json
```

```python
with open('notionDB.json') as jsonfile:
    data = json.load(jsonfile)
    df = pd.DataFrame(data)
```

## Preprocessing database data

```python
df = df[['id', 'properties', 'cover']]
df
```

```python
cover_df = pd.read_json(df['cover'].to_json()).T
cover_df = pd.read_json(cover_df['external'].to_json()).T
cover_df.reset_index(inplace=True)
```

```python
df['cover'] = cover_df['url']
```

```python
properties_df = pd.read_json(df['properties'].to_json()).T
```

```python
url_df = pd.read_json(properties_df['URL'].to_json()).T
url_df = url_df.drop(['id', 'type'], axis = 1)
url_df.reset_index(inplace=True)
```

```python
df['url'] = url_df['url']
```

```python
titles_raw = pd.read_json(properties_df['Title'].to_json()).T
titles_list = []

for i in range(len(titles_raw)):
  titles_list.append(titles_raw['title'][i][0]['text']['content'])

titles_df = pd.DataFrame(titles_list)
```

```python
df['title'] = titles_df
```

```python
ratings_raw = pd.read_json(properties_df['Ratings'].to_json()).T
ratings_raw = ratings_raw.fillna(value=np.nan)
ratings_list = []

for i in range(len(ratings_raw)):
  item = ratings_raw['select'][i]
  if type(item) == float:
    ratings_list.append(0)
  else:
    ratings_list.append(len(item['name']))

ratings_df = pd.DataFrame(ratings_list)
```

```python
df['ratings'] = ratings_df
```

```python
df = df.drop('properties', axis = 1)
```

```python
df
```

```python
# Remove all characters that are not alphabet, numbers, and whitespace
df['mod_title'] = df['title'].str.replace("[^a-zA-Z0-9 ]", "", regex = True)

# Change to small letter
df['mod_title'] = df['mod_title'].str.lower()

# Replace a string of non-whitespace characters to a whitespace
df['mod_title'] = df['mod_title'].str.replace('\s+', ' ', regex= True)
```

```python
df = df[df['mod_title'].str.len() > 0]
```

```python
df.to_json("books_titles.json")
```

## Building Search Engine

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

import re
```

```python
vectorizer = TfidfVectorizer()
tfidf = vectorizer.fit_transform(df['mod_title'])
```

```python
def make_clickable(val):
  return '<a target = "_blank" href = "{}">Link</a>'.format(val)
```

```python
def show_image(val):
  return '<img src = "{}" width = 50></img>'.format(val)
```

```python
def search(query, vectorizer):
  processed = re.sub("[^a-zA-Z0-9 ]", "", query.lower())
  query_vec = vectorizer.transform([processed])
  similarity = cosine_similarity(query_vec, tfidf).flatten()

  # Getting the index of the 10 highest similarity element
  indices = np.argpartition(similarity, -10)[-10:]
  results = df.iloc[indices]
  results = results.sort_values('ratings', ascending = False)

  return results.style.format({'url': make_clickable, 'cover': show_image})
```

```python
search("Investing 101", vectorizer)
```
